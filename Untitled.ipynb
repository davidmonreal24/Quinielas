```python
import pandas as pd

# Cargar los datasets
schedule = pd.read_parquet("data/processed/schedule_xg.parquet")
players  = pd.read_parquet("data/processed/player_season_stats.parquet")

print("=== SCHEDULE XG ===")
print(f"Forma: {schedule.shape}")
print(f"\nColumnas:\n{list(schedule.columns)}")
print(f"\nPrimeras filas:")
schedule.head(3)
```

```python
print("=== VALORES NULOS ===")
print(schedule.isnull().sum())

print("\n=== LIGAS Y TEMPORADAS ===")
print(schedule.groupby(["league", "season"]).size().to_string())

print("\n=== DISTRIBUCIÓN DE RESULTADOS ===")
schedule["resultado"] = schedule.apply(
    lambda r: "Local" if r.home_goals > r.away_goals 
    else ("Visitante" if r.home_goals < r.away_goals else "Empate"), axis=1
)
print(schedule["resultado"].value_counts())
print(f"\n% Local:     {schedule['resultado'].eq('Local').mean()*100:.1f}%")
print(f"% Empate:    {schedule['resultado'].eq('Empate').mean()*100:.1f}%")
print(f"% Visitante: {schedule['resultado'].eq('Visitante').mean()*100:.1f}%")

print("\n=== ESTADÍSTICAS DE xG ===")
print(schedule[["home_xg", "away_xg", "home_goals", "away_goals"]].describe().round(2))
```

```python
print("=== PLAYERS DATASET ===")
print(f"Forma: {players.shape}")
print(f"\nColumnas:\n{list(players.columns)}")

print("\n=== VALORES NULOS ===")
print(players.isnull().sum())

print("\n=== MUESTRA DE DATOS ===")
players.head(3)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1. Distribución de resultados
schedule["resultado"].value_counts().plot(
    kind="bar", ax=axes[0], color=["#2ecc71","#e74c3c","#95a9a6"],
    title="Distribución de Resultados"
)

# 2. xG Local vs xG Visitante
axes[1].scatter(schedule["home_xg"], schedule["away_xg"], alpha=0.2, s=10, color="#3498db")
axes[1].set_xlabel("xG Local")
axes[1].set_ylabel("xG Visitante")
axes[1].set_title("xG Local vs Visitante")

# 3. Goles reales vs xG local
axes[2].scatter(schedule["home_xg"], schedule["home_goals"], alpha=0.2, s=10, color="#e67e22")
axes[2].set_xlabel("xG Local")
axes[2].set_ylabel("Goles Reales Local")
axes[2].set_title("xG vs Goles Reales")

plt.tight_layout()
plt.savefig("data/exploracion.png", dpi=150)
plt.show()
print("✅ Gráfica guardada en data/exploracion.png")
```
